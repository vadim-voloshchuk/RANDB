```python
#   
import pandas as pd
import matplotlib.pyplot as plt

#  1:    CSV
file_path = 'training_log_normal_reward_level2_NOWAY.csv'  #     CSV-
data = pd.read_csv(file_path)

#  2:  
plt.figure(figsize=(14, 10))

#  
plt.subplot(3, 2, 1)
plt.plot(data['Episode'], data['Reward'], label='Reward', color='blue')
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.title('Rewards per Episode')
plt.grid(True)

#  
plt.subplot(3, 2, 2)
plt.plot(data['Episode'], data['Wins'], label='Wins', color='green')
plt.xlabel('Episode')
plt.ylabel('Wins')
plt.title('Wins per Episode')
plt.grid(True)

#  
plt.subplot(3, 2, 3)
plt.plot(data['Episode'], data['Losses'], label='Losses', color='red')
plt.xlabel('Episode')
plt.ylabel('Losses')
plt.title('Losses per Episode')
plt.grid(True)

#  Epsilon
plt.subplot(3, 2, 4)
plt.plot(data['Episode'], data['Epsilon'], label='Epsilon', color='orange')
plt.xlabel('Episode')
plt.ylabel('Epsilon')
plt.title('Epsilon Decay per Episode')
plt.grid(True)

#  
plt.tight_layout()
plt.show()

#  3: 
average_reward = data['Reward'].mean()
total_wins = data['Wins'].max()
total_losses = data['Losses'].max()
final_epsilon = data['Epsilon'].iloc[-1]

print(":")
print(f" : {average_reward:.2f}")
print(f"  : {total_wins}")
print(f"  : {total_losses}")
print(f"  Epsilon: {final_epsilon:.4f}")
```
