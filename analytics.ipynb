```python
# Импорт необходимых библиотек
import pandas as pd
import matplotlib.pyplot as plt

# Шаг 1: Загрузка данных из CSV
file_path = 'training_log_normal_reward_level2_NOWAY.csv'  # Укажите путь к вашему CSV-файлу
data = pd.read_csv(file_path)

# Шаг 2: Визуализация данных
plt.figure(figsize=(14, 10))

# График наград
plt.subplot(3, 2, 1)
plt.plot(data['Episode'], data['Reward'], label='Reward', color='blue')
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.title('Rewards per Episode')
plt.grid(True)

# График выигрышей
plt.subplot(3, 2, 2)
plt.plot(data['Episode'], data['Wins'], label='Wins', color='green')
plt.xlabel('Episode')
plt.ylabel('Wins')
plt.title('Wins per Episode')
plt.grid(True)

# График проигрышей
plt.subplot(3, 2, 3)
plt.plot(data['Episode'], data['Losses'], label='Losses', color='red')
plt.xlabel('Episode')
plt.ylabel('Losses')
plt.title('Losses per Episode')
plt.grid(True)

# График Epsilon
plt.subplot(3, 2, 4)
plt.plot(data['Episode'], data['Epsilon'], label='Epsilon', color='orange')
plt.xlabel('Episode')
plt.ylabel('Epsilon')
plt.title('Epsilon Decay per Episode')
plt.grid(True)

# Общий вывод
plt.tight_layout()
plt.show()

# Шаг 3: Выводы
average_reward = data['Reward'].mean()
total_wins = data['Wins'].max()
total_losses = data['Losses'].max()
final_epsilon = data['Epsilon'].iloc[-1]

print("Выводы:")
print(f"Среднее вознаграждение: {average_reward:.2f}")
print(f"Общее количество выигрышей: {total_wins}")
print(f"Общее количество проигрышей: {total_losses}")
print(f"Конечное значение Epsilon: {final_epsilon:.4f}")
```
